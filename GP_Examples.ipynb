```python
import pylab as pl
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel, WhiteKernel, ExpSineSquared
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(302)
t = [1., 3., 5., 6., 7., 8.]
t_ = np.atleast_2d(t).T
t_
```

```python
def f(x):
  np.random.seed(302)
  x = x.ravel()
  return np.random.rand(len(x))
```

```python
y = f(t_)
y_err = np.abs(y)/2
```

```python
pl.plot(t_.ravel(), y)
```

```python
plt.errorbar(t_.ravel(), y, yerr=y_err, fmt='o')
plt.xlabel("t")
plt.ylabel("y")
plt.title("y vs t with error bars")
plt.show()
```
